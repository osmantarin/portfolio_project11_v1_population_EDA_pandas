```python
# We will be examining data containing population growth globally since 1970 

# Our use case: Which continent has grown the fastest? Are there certain continents that have grown faster than others? 
```

```python
# We first import the tools necessary to complete our data analysis 
        
    # We will utilize the Pandas library to analyze our data
    
    # We will utilize seaborn and the Matplotlib visualization library to to plot our data
    
   

import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
# We import our data from the csv file into a dataframe

df = pd.read_csv(r"C:\Users\itsno\Desktop\Portfolio Projects\portfolio_project11_v1_EDA_pandas\csv_data\world_population.csv")
                 
df
```

```python
# We format and standartdize our entire dataset to within two decimal places 

pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
# We review key high level information about our dataset 

df.info()
```

```python
# We review key high level information about our columns 

df.describe()

```

```python
# We review our dataset for missing values 

    # The results below show that we defintely have missing data

df.isnull().sum()
```

```python
# We review our dataset for unique values 

df.nunique()
```

```python
# We continue previewing our dataset, in this case, sorting by World Population Percentage

df.sort_values(by="World Population Percentage", ascending=False).head(10)

```

```python
# We compare columns within the dataset to see how closely correlated they are

df.corr(numeric_only=True)
```

```python
# We visualize the above correlations using a heat map

    # From the heat map below, it seems that the Area is highly correlated with the population size

sns.heatmap(df.corr(numeric_only=True), annot = True)

plt.rcParams['figure.figsize'] = (20,7)

plt.show
```

```python
# We continue to explore the dataset, in this case grouping the data by Continent and sorting by the population value in 2022

    # Note, we assign this grouping to a dataframe in order to visualize our data
    
df2 = df.groupby ('Continent').mean(numeric_only=True).sort_values(by="2022 Population", ascending=False)
```

```python
# We continue to explore the dataset, in this case, reviewing data only for the Continent of Oceania 

df[df['Continent'].str.contains('Oceania')].head(10)
```

```python
# Our initial attempt to visualize the data results in a plot that is not readily intuitive due to location of the axis

    # We need to transpose our dataset in order for continents to appear on the y-axis and population on the x-axis

df2.plot()
```

```python
# We transpose the data in order to improve our visualization and assign it to a new dataframe 

df3 = df2.transpose()

df3
```

```python
# Our plot is still not readily intuitive as we have unnecessary columns in the plot

    # We must select the appropriate columns for our plot

df3.plot()

```

```python
# We quickly review the columns we feel will be relevant to the plot

df.columns
```

```python
# We define a new dataframe with the relevant columns and transpose this dataframe 

df4 = df.groupby ('Continent')[['1970 Population', '1980 Population', '1990 Population',
       '2000 Population', '2010 Population', '2015 Population',
       '2020 Population', '2022 Population']].mean(numeric_only=True).sort_values(by="2022 Population", ascending=False)

df5 = df4.transpose()

df5
```

```python
# The plot is now visually intuitive

# At a glance, we can see Asia is the continent with the largest population growth since 1990

# We have answered our initial use case question

df5.plot()
```
